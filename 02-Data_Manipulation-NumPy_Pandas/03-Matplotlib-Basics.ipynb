# #Multiple Figures

first line represents the whole size determined by 12 and 8 all 4 

second line sub plot is the position of the graphs so 2x2

tight layout makes the layout/spacing look better and more professional but doesnt hurt anything

```python
plt.figure(figsize=(12,8))
# 2 indicates the amount of rows, the subpoint, the 3 is how many graphs are in each row,
# the 1 indicates the order of each graph in each row differentiating each graph from each row
plt.subplot(2,3,1) 
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,3,1)')

plt.subplot(2,3,2)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,3,2)')

plt.subplot(2,3,3)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,3,3)')

plt.subplot(2,3,4)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,3,4)')

plt.subplot(2,3,5)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,3,5)')

plt.subplot(2,3,6)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,3,6)')


plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('nbadatabase_24_25.csv')
df
```

```python
x = df ['MP']
y = df ['PTS']
#Create scatter plot
plt.figure(figsize=(28, 16))
plt.scatter(x,y, alpha=0.3)

#Add labels and title
plt.xlabel("Minutes Played")
plt.ylabel("Points Scored")
plt.title("Scatter Plot to Represent the Relationship Between Minutes Played and Points SCORED(Using NumPy)")

# Show the plot
plt.show()
```

```python
# As players spend more time on the court, they generally score more points.
# The cloud of dots gets higher as it moves to the right, showing how opportunity increases production.
# This scatter plot represents the relationship between minutes played and points scored in the NBA, where each
# dot represents a players performance in a single game. The graph shows a strong positive correlation, 
# meaning that as minutes played increase, points scored generally increase as well. Players who play fewer
# than 10 minutes typically score very few points, while players who are on the court for 3045 minutes tend
# to score much higher, with some reaching 4060 points. However, the wide spread of points at higher minutes
# shows that playing time alone does not guarantee high scoring, as skill level, efficiency, and role on the
# team also affect performance. Overall, the graph shows that minutes played create more scoring opportunities,
# but individual ability ultimately determines how effectively those opportunities are used.
# 
```

```python
# The graph proves that to be a top scorer, you must first earn the coach's trust to stay on the floor.
#  However, once you have those minutes, your skill level (the Y-axis height) determines if you become a
#  Role Player or a Superstar.
```

```python
# In this graph my objective is to determine a players efficentcy minutes played versus points scored
# However, in order to fully obtain historical context you need many and balanced perspectives comparing 
# everything in relation to efficenty(so multiple graphs because one graph cant capture all information). 
# This graph single handedly cannot accuratly represent the efficentcy and in order to determine this you must 
# have multiple comparissions betweeen efficentcy and other box stats in order to obatain a balanced and well 
# rounded perspective upon player efficentcy and minutes but you can also break it down into different stats
```

```python
# Definng efficentcy between points scored and time
df['Efficiency'] = df['PTS'] / df['MP']
```
