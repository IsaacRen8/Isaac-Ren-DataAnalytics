1. Data Cleaning  
Data cleaning is the process of fixing or removing incorrect, missing, or duplicate data so it can be used properly.  
Example: Removing duplicate student records or filling in missing values in a dataset.

2. Feature Engineering  
Feature engineering is creating or modifying input variables (features) to improve how well a model works.  
Example: Converting a timestamp into useful features like hour, day, or month.

3. Machine Learning  
Machine learning is when computers learn patterns from data and use them to make predictions or decisions.  
Example: A recommendation system suggesting videos based on what you watched before.

4. Supervised Learning  
Supervised learning is when a type of machine is trained using labeled data, meaning the correct answers are given.
Example: Training a model with emails labeled as spam or not spam.

5. Unsupervised Learning  
Unsupervised learning is when a model finds patterns in data without any labels. Nothing to tell right and wrong only determined through patterns always has errors because you are sorting without knowing what is what  
Example: Grouping users based on behavior without being told the groups.

6. Classification  
Classification is when a model predicts a category or label. You know / are aware of labells. Type of supervised learning 
Example: Deciding if an email is spam or not spam.

7. Regression  
Regression is when a model predicts a number.  
Example: Predicting the price of a house.

8. Clustering  
Clustering is grouping similar data points together.  
Example: Grouping customers based on their buying habits.

9. Dimensional Reduction  
Dimensional reduction is reducing the number of features while keeping the important information.   
Example: Reducing 100 features to 10 to make a model faster and easier to use.
