# Random

## rand

## Creates an array of the given shape and populates it with random samples froma. uniform distribution over [0,1)

```python
np.random.rand(2) # Rand generates only a number between[0,1]
```

```python
np.random.rand(4)
```

```python
np.random.rand(5,5) # works with arrays
```

## Randint

### Returns random integers where the lowest number is inclusive but the highest exclusive

### Return random integers from low (inclusive) to high (exclusive)

```python
np.random.randint(1,100) # generate a random number from lower bound to upper bound
                         # lower bound number is included while upper number is excluded
```

```python
np.random.randint(1,100,10) # third number is for how many numbers in the array print 
```

## seed

### Can be used to set the random state, so that the same "random" resultes can be reproduced

```python
np.random.seed(67)
np.random.rand(4)
```

```python
np.random.seed(67)
np.random.randint(1,100,4)
```
