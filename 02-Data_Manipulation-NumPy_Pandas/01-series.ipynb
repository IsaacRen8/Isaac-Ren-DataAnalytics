# Creating a series through a list

```python
mydata=[1776,1867,1621] # mydata = list
myseries=pd.Series(data=mydata) #on the right hand of = side is important
                                #pd.Series=Pandasseries
                                #pd.Series(data=mydata) is pd's Series
                                #Series created from data
                                #MySeries series

myseries          
                                #dtype is short for data type
                                # int32 example of refrence range of integer(-100,100).  2^32=int32
                                # int64 example of refrence range of integer(-1000,1000). 2^64=int64
```

# Create series by customize index


```python
myindex=['USA','Canada','Mexico']
myseries2= pd.Series (data = mydata, index = myindex)#adding names for the index/columm
myseries2
```

# Series Operation

```python
q1= {'Japan':80,
    'China':450,
    'USA':300,
    'India':100,}
q2= {'Brazil':70,
    'China':67,
    'USA':87,
    'India':88,}

sale_q1=pd.Series(q1)
sale_q2=pd.Series(q2)
q1
q2


```

## Call Series

```python
sale_q1['Japan']#Both ways work but encouraged to use this one as it avoids calling the index number
```

```python
sale_q1[0]#Works but method above is encouraged
```

## Wrong Way of Calling

```python
# ALL WERE INCORRECT
sale_q1['France']# Non Existent Name
sale_q2['USA '] #incorrect bc of extra space
sale_q1['japan'] # Case sensitive/undercase
```
