## Basic Matplotlib Commands

```python
plt.plot(x,y)
plt. xlabel('X Axis Title Here')
plt. ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show() # Required for non-jupyter users, but also removes out[] info
```

```python
plt.plot(x,y)
plt. xlabel('X Axis Title Here')
plt. ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.xlim(0,6) # Lower Limit, Upper Limit
plt.ylim(0,12) # Lower Limit, Upper Limit
plt.show() # Required for non-jupyter users, but also removes out[] info
```

### Figure Size

```python
plt.figure(figsize=(12,8))

plt.plot(x,y)
plt. xlabel('X Axis Title Here')
plt. ylabel('Y Axis Title Here')
plt.title('String Title Here')

plt.show() # Required for non-jupyter users, but also removes out[] info
```

# #Multiple Figures

first line represents the whole size determined by 12 and 8 all 4 

second line sub plot is the position of the graphs so 2x2

tight layout makes the layout/spacing look better and more professional but doesnt hurt anything

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,2,1)')

plt.subplot(2,2,2)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,2,2)')

plt.subplot(2,2,3)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,2,3)')

plt.subplot(2,2,4)
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('(2,2,4)')

plt.tight_layout()
plt.show()
```
