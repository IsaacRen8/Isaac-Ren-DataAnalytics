## Function with input, no return

```python
def sum_num(num1,num2):
    print (f" num1 = {num1}")
    print (f" num2 = {num2}")
    result=num1+num2
    print (result)

print("positionl example1:")
sum_num(3,4)
print("-------------------\n")

print("positionl example2:")
sum_num(10,5) # when you call the function the 3 is passing to num1/3,4 are equivilant to num1 and num2
            # (3,4) are called parameters
            # spots are defined because they are in a "positional"
print("-------------------\n")

print("argumentqal example1:")
# sum_num(3,2)
sum_num(num2=3,num1=2)
```
